```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
print(sns.get_dataset_names())
```

```python
#load and read dataset
df = sns.load_dataset("penguins")
df[:4]
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
#df.dropna()
df1=df.dropna()
df1
```

```python
df1.isna().sum()
```

```python
df1["sex"][:4]
```

```python
#encoding
ohe =OneHotEncoder(sparse_output=False, dtype="int")

gender_ohe = ohe.fit_transform(df1[["sex"]])
pd.DataFrame(gender_ohe, columns=["Sex_Female", "Sex_Male"])[:6]
```

```python
# One-hot encode island
island_encoder = OneHotEncoder(drop='first', sparse_output=False, dtype="int")
island_encoded = island_encoder.fit_transform(df1[['island']])

# Convert to DataFrame
island_df = pd.DataFrame(island_encoded, columns=island_encoder.get_feature_names_out(['island']))
island_df
```

```python
#example of label encoder
lec = LabelEncoder()
species_lec = lec.fit_transform(df["species"])
species_lec
```

```python
#ONE encode all cat columns
cat_coln = df.select_dtypes("category")
cat_coln
```

```python
#encode
encoded_col= pd.get_dummies(cat_coln, dtype= 'int', drop_first=True)
encoded_col
```
