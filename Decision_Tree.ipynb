**Decision Tree**

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
 #sklearn libraries
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve,auc, roc_auc_score,auc
```

```python
# Load Iris dataset
data = sns.load_dataset('iris')
data[:5]
```

```python
#checking  missing values
data.isna().sum()
```

```python
#remove duplicates
data.drop_duplicates(inplace=True)

data.duplicated().sum()#.any()
```

```python
data["species"].value_counts()
```

```python
#convert label to numerical values
data["species"]= data["species"].map({"setosa":0, "versicolor":1, "virginica":2})
data["species"] = data["species"].astype("int")
```

```python
#separate features from label
X = data.drop("species", axis=1)
y = data.species
```

```python
#dividing train and test sets
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size=.3, random_state=42)
```

```python
#checking shape
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#1 call and train the model (ID3 MODEL)
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
#checking training score
model.score(x_train, y_train)
```

```python
#2. predict
y_pred = model.predict(x_test)
#y_pred
result= {"Actual": y_test, "Predicted": y_pred}
pd.DataFrame(result)[:5]
```

```python
#3. model score
accuracy_score(y_test, y_pred)
```

```python
# check confusion
conf = confusion_matrix(y_test, y_pred)
conf
```

```python
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(10,10))
tree.plot_tree(model, filled=True); # we can also visualize using plot_tree(model, filled=True)
```

## Pruning the model to avoid overfitting

```python
#prune the tree
model2 = DecisionTreeClassifier(max_depth=3)
model2.fit(x_train, y_train)
#predict
y_pred = model2.predict(x_test)
#model_score
accuracy_score(y_test, y_pred)
```

```python
#visualize
plt.figure(figsize=(10,10))
plot_tree(model2,filled=True);

```

**Create CART model**

```python
#CART model
model3 = DecisionTreeClassifier(criterion="entropy", max_depth= 3)
model3.fit(x_train, y_train)
#2.predict
y_pred = model3.predict(x_test)

#3. model score
accuracy_score(y_test, y_pred)
```

```python
#visualize
plt.figure(figsize=(10,10))
tree.plot_tree(model3, filled=True);
```

**Get Feature Importance**

we need to identify the features that are important in helping with the model prediction.

```python
#get features score
feature = model3.feature_importances_
feature
```

```python
#get features names
names = model3.feature_names_in_
names
```

```python
#combine
result = {"name":names, "feature_imp":feature}
fi = pd.DataFrame(result)
fi
```

```python
# sort the features in descending order
t = fi[["name", "feature_imp"]].sort_values(ascending=False, by= "feature_imp")
t
```

```python
#visualizing feature importance
sns.barplot(x="name", y="feature_imp", data=t)
plt.xticks(rotation=45, ha = "right");
```

```python
#another way of representing feature importance
sns.barplot(y="name", x="feature_imp", data=t)
plt.tight_layout();
```

Observations: Petal lenght and petal width are most vital features in predicting the species of the flower.
